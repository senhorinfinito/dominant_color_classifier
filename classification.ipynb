```python
import os
import cv2
import json
import shutil
import numpy as np
import tensorflow as tf
```

```python
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
tf.executing_eagerly()
```

```python
def mkdir():
    if not os.path.exists('./output'):
        os.makedirs('./output')
    if not os.path.exists('./output/red'):
        os.makedirs('./output/red')
    if not os.path.exists('./output/green'):
        os.makedirs('./output/green')
    if not os.path.exists('./output/blue'):
        os.makedirs('./output/blue')
    if not os.path.exists('./output/orange'):
        os.makedirs('./output/orange')
    if not os.path.exists('./output/yellow'):
        os.makedirs('./output/yellow')
    if not os.path.exists('./output/pink'):
        os.makedirs('./output/pink')
    if not os.path.exists('./output/purple'):
        os.makedirs('./output/purple')
    if not os.path.exists('./output/brown'):
        os.makedirs('./output/brown')
    if not os.path.exists('./output/grey'):
        os.makedirs('./output/grey')
    if not os.path.exists('./output/other'):
        os.makedirs('./output/other')
```

```python
def preProcessing(name):
    img = cv2.imread(name) / 255  # Normalized image the required entry interval is [0,1]
    w, h, c = img.shape
    size = w * h
    img = img[..., ::-1]  # This one is funny BGR->RGB
    img = cv2.resize(img, (10, 10))  # Resize to accelerate speed
    img = img.reshape(-1, 3)  # Shape 2D image into 1D vector
    return img, size
```

```python
def compute(img):
    colors = tf.constant(img, dtype=tf.float32)
    model = tf.keras.models.model_from_json(json.load(open("model.json"))["model"], custom_objects={})
    model.load_weights("model_weights.h5")
    predictions = model.predict(colors, batch_size=32, verbose=0)
    predictions = tf.one_hot(np.argmax(predictions, 1), 9)
    # Sum along the column, each entry indicates no of pixels
    res = tf.reduce_sum(predictions, axis= 0 ).numpy()
    # Threshold is 0.5 (accuracy is 96%) change threshold may cause accuracy decrease
    if res[0] / (sum(res[:-1]) + 1) > 0.75:
        return "red"
    elif res[1] / (sum(res[:-1]) + 1) > 0.75:
        return "green"
    elif res[2] / (sum(res[:-1]) + 1) > 0.75:
        return "blue"
    elif res[3] / (sum(res[:-1]) + 1) > 0.75:
        return "orange"
    elif res[4] / (sum(res[:-1]) + 1) > 0.75:
        return "yellow"
    elif res[5] / (sum(res[:-1]) + 1) > 0.75:
        return "pink"
    elif res[6] / (sum(res[:-1]) + 1) > 0.75:
        return "purple"
    elif res[7] / (sum(res[:-1]) + 1) > 0.75:
        return "brown"
    elif res[8] / (sum(res[:-1]) + 1) > 0.75:
        return "grey"
    else:
        return "other"
```

```python
def copy(name, full_name, size, label):
    if label == 'red':
        shutil.copy(full_name, "./output/red")
        os.rename('./output/red/' + name, './output/red/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'green':
        shutil.copy(full_name, "./output/green")
        os.rename('./output/green/' + name, './output/green/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'blue':
        shutil.copy(full_name, "./output/blue")
        os.rename('./output/blue/' + name, './output/blue/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'orange':
        shutil.copy(full_name, "./output/orange")
        os.rename('./output/orange/' + name, './output/orange/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'yellow':
        shutil.copy(full_name, "./output/yellow")
        os.rename('./output/yellow/' + name, './output/yellow/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'pink':
        shutil.copy(full_name, "./output/pink")
        os.rename('./output/pink/' + name, './output/pink/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'purple':
        shutil.copy(full_name, "./output/purple")
        os.rename('./output/purple/' + name, './output/purple/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'brown':
        shutil.copy(full_name, "./output/brown")
        os.rename('./output/brown/' + name, './output/brown/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'grey':
        shutil.copy(full_name, "./output/grey")
        os.rename('./output/grey/' + name, './output/grey/' + str(size) + str(name.split('.')[0]) + '.jpg')
        
    if label == 'other':
        shutil.copy(full_name, "./output/other")
        os.rename('./output/other/' + name, './output/other/' + str(size) + str(name.split('.')[0]) + '.jpg')
```

```python
def main():
    files = os.listdir('./input/')
    for name in files:
        mkdir()
        full_name = './input/' + name
        img, size = preProcessing(full_name)
        label = compute(img)
        copy(name, full_name, size, label)
    sort()
```

```python
main()
```
